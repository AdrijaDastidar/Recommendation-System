**Neo4j Query Execution Helper Function** - A Neo4j driver is created to handle database connections. The run_query function is defined to execute Cypher queries in the database.

```python
from neo4j import GraphDatabase
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))
def run_query(query, parameters=None):
    with driver.session() as session:
        session.run(query, parameters=parameters)
```

**Data Loading Functions**
1. Load nodes to the function to neo4j.
  1. Load Movies
  2. Load Genres
  3. Load Ratings

  
2. Creating Similarity Relationships
  1. User Similarity
  2. Movie Similarity by Genre
  3. Movie Similarity by Year


```python
def load_movies(movies_rating):
    # Load movie nodes with properties: id, title, year_group, rating_mean
    for _, row in movies_rating.iterrows():
        query = """
        MERGE (m:Movie {id: $id})
        SET m.title = $title,
            m.year_group = $year_group,
            m.rating_mean = $rating_mean
        """
        parameters = {
            "id": row['id'],
            "title": row['title'],
            "year_group": int(row['year_group']),
            "rating_mean": float(row['rating_mean'])
        }
        run_query(query, parameters)
    print("Movies with titles loaded successfully.")



def load_genres(movies_genre):
    genre_columns = ['Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary',
                     'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
                     'Sci-Fi', 'Thriller', 'War', 'Western']

    for _, row in movies_genre.iterrows():
        for genre in genre_columns:
            if row[genre] == 1:
                query = """
                MERGE (m:Movie {id: $id})
                MERGE (g:Genre {name: $genre})
                MERGE (m)-[:HAS_GENRE]->(g)
                """
                parameters = {"id": row['id'], "genre": genre}
                run_query(query, parameters)
    print("Genres and genre relationships loaded successfully.")



def load_ratings(filtered_ratings):
    for _, row in filtered_ratings.iterrows():
        query = """
        MERGE (u:User {id: $user_id})
        MERGE (m:Movie {id: $movie_id})
        CREATE (u)-[:RATES {rating: $rating, timestamp: $timestamp}]->(m)
        """
        parameters = {
            "user_id": row['user_id'],
            "movie_id": row['movie_id'],
            "rating": float(row['ratings']),
            "timestamp": int(row['timestamp'])
        }
        run_query(query, parameters)
    print("Ratings loaded successfully.")
```

```python
load_movies(movies_rating)
load_genres(movies_genre)
load_ratings(filtered_ratings)
```

```python
def create_user_similarity_relationship(graph, rating_threshold=0.5):
    query = f"""
    MATCH (u1:User)-[r1:RATES]->(m:Movie)<-[r2:RATES]-(u2:User)
    WHERE u1 <> u2 AND
          abs(r1.rating - r2.rating) <= {rating_threshold}
    MERGE (u1)-[:SIMILAR_TO_Ratings]->(u2)
    """
    try:
        graph.query(query)
        print("User similarity relationships created with constraints.")
    except Exception as e:
        print("An error occurred while creating user similarity relationships:", e)



def create_movie_similarity_relationship_by_genre(graph, genre_threshold=2, rating_diff_threshold=2):
    query = f"""
    MATCH (m1:Movie)-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(m2:Movie)
    WHERE m1 <> m2 AND
          abs(m1.rating_mean - m2.rating_mean) <= {rating_diff_threshold}
    WITH m1, m2, COUNT(g) AS common_genres
    WHERE common_genres >= {genre_threshold}
    MERGE (m1)-[:SIMILAR_TO_GENRE {{score: common_genres}}]->(m2)
    """
    try:
        graph.query(query)
        print("Movie similarity relationships based on genre created successfully.")
    except Exception as e:
        print("An error occurred while creating movie similarity relationships by genre:", e)

def create_movie_similarity_relationship_by_year(graph, year_threshold=1, genre_threshold=2):
    query = f"""
    MATCH (m1:Movie)-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(m2:Movie)
    WHERE m1 <> m2 AND
          abs(m1.year_group - m2.year_group) <= {year_threshold}
    WITH m1, m2, COUNT(g) AS common_genres
    WHERE common_genres >= {genre_threshold}
    MERGE (m1)-[:SIMILAR_TO_YEAR {{score: common_genres}}]->(m2)
    """
    try:
        graph.query(query)
        print("Movie similarity relationships based on year created successfully.")
    except Exception as e:
        print("An error occurred while creating movie similarity relationships by year:", e)

```

**Recommendations**
1. Based on Similar Users
2. Based on Genre and Year

```python
def recommend_movies_based_on_similar_users(graph, user_id, top_n=10):
    query = f"""
    MATCH (u:User {{id: {user_id}}})-[:SIMILAR_TO_Ratings]-(similar_user:User)-[r:RATES]->(m:Movie)
    WHERE NOT (u)-[:RATES]->(m)
    RETURN m.title AS movie_title, COLLECT(DISTINCT similar_user.id) AS similar_users, AVG(r.rating) AS average_rating
    ORDER BY average_rating DESC
    LIMIT {top_n}
    """

    try:
        recommendations = graph.query(query)
        print(f"Movies recommended based on similar users for user {user_id}:")
        if not recommendations:
            print("No recommendations found.")
        for rec in recommendations:
            print(f"Title: {rec['movie_title']} \t| Average Rating: {rec['average_rating']} \t| Similar Users: {rec['similar_users']}")
    except Exception as e:
        print("An error occurred while fetching recommendations:", e)
```

```python
user_id = 1
recommend_movies_based_on_similar_users(graph, user_id)
```

```python
def recommend_movies_based_on_genre_and_year(graph, movie_id, genre_threshold=1, top_n=10):
    query = f"""
    MATCH (m1:Movie {{id: {movie_id}}})-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(m2:Movie)
    WHERE m1 <> m2
    WITH m1, m2, COUNT(g) AS genre_count
    WHERE genre_count >= {genre_threshold} AND abs(m1.year_group - m2.year_group) <= 1
    RETURN m2.title AS movie_title
    ORDER BY m2.title
    LIMIT {top_n}
    """

    try:
        recommendations = graph.query(query)
        print(f"Movies recommended based on genre and year for movie {movie_id}:")
        for rec in recommendations:
            print(f"Title: {rec['movie_title']}")
    except Exception as e:
        print("An error occurred while fetching recommendations:", e)
```

Score Calculation

```python
def calculate_metrics(recommended_movies, rated_movies_set, k=10):
    # Get the top k recommended movies
    top_k_recommendations = recommended_movies[:k]

    # Calculate Precision@k
    true_positives = len(set(top_k_recommendations) & rated_movies_set)
    precision_at_k = true_positives / k if k > 0 else 0.0

    # Calculate Recall@k
    recall_at_k = true_positives / len(rated_movies_set) if len(rated_movies_set) > 0 else 0.0

    # Calculate F1-score@k
    if precision_at_k + recall_at_k > 0:
        f1_at_k = 2 * (precision_at_k * recall_at_k) / (precision_at_k + recall_at_k)
    else:
        f1_at_k = 0.0

    return precision_at_k, recall_at_k, f1_at_k
```

```python
def recommend_movies_based_on_similar_users(graph, user_id, top_n=50):
    query = f"""
    MATCH (u:User {{id: {user_id}}})-[:SIMILAR_TO_Ratings]-(similar_user:User)-[r:RATES]->(m:Movie)
    RETURN m.id AS movie_Id
    LIMIT {top_n}
    """

    try:
        recommendations = graph.query(query)

        if not recommendations:
            print("No recommendations found.")
        recommended_movies = [rec['movie_Id'] for rec in recommendations]
        print("Recommended movies:", recommended_movies)


        rated_movies_query = f"""
        MATCH (u:User {{id: {user_id}}})-[:RATES]->(m:Movie)
        RETURN m.id AS movie_Id
        """
        rated_movies = graph.query(rated_movies_query)
        print(rated_movies)

        rated_movies_set = {rec['movie_Id'] for rec in rated_movies}

        # Calculate Precision, Recall, F1-score
        precision, recall, f1 = calculate_metrics(recommended_movies, rated_movies_set, top_n)

        # Output Precision, Recall, F1-Score
        print(f"Precision@{top_n}: {precision}")
        print(f"Recall@{top_n}: {recall}")
        print(f"F1-score@{top_n}: {f1}")

    except Exception as e:
        print("An error occurred while fetching recommendations:", e)
```
